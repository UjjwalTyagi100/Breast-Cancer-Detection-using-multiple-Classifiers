### Lower case message data

My next step in preprocessing is to set all of the characters to lower case by chaining the `.str()` method to convert the pandas Series so that the `.lower()` method can be applied to lower case each string in the column.

```python
dataset['Message_clean'] = dataset['Message'].str.lower()

dataset.head()
```

### Remove punctuation

Now that all of the characters are in lower case, I will use the same approach for removing punctuation. I will chain the `.str()` method with the `.replace()` method to remove any punctuation.

```python
dataset['Message_clean'] = dataset['Message_clean'].str.replace('[^\w\s]' , '')

''' '[^\w\s]': This is a regular expression (regex) pattern.
[^\w\s]: This pattern matches any character that is not a word character (\w) or whitespace character (\s).
\w: Matches any alphanumeric character (letters and digits) and underscores.
\s: Matches any whitespace character (spaces, tabs, newlines).
Therefore, [^\w\s] matches any non-alphanumeric and non-whitespace characters, which typically includes punctuation and special characters.

`''`: This is the replacement string. Here, it is an empty string, meaning that characters matching the regex pattern will be removed from the text.
'''

dataset.head()
```

### Remove stopwords

### Remove URLs

For some datasets, URL removal is a useful component of the text preprocessing phase, so I'll remove them here using regular expressions.

```python
import re # Here, 're' means 'Regular Expression'
```

```python
# Creating a function to remove url from the message data
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+') # This line creates a 'regex' means 'Regular Expression' pattern that matches the urls in the given text
    return url_pattern.sub(r'', text)
```

```python
# Removing the urls
dataset['Message_clean'] = dataset['Message_clean'].apply(lambda text: remove_urls(text))
# View changes
dataset.head()
```

### Remove HTMLs

Similarly, I will remove the HTML tags using the re library's `.compile()` method.

```python
# Creating a function to remove HTML tags
def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r"", text)
```

```python
# Removing the HTML tags from the message data
dataset['Message_clean'] = dataset['Message_clean'].apply(lambda text: remove_html(text))

# view changes
dataset.head()
```

### Sentiment Analysis

Now that I have conducted some light EDA (Exploratory Data Analysis), I will move on to sentiment analysis. There are several approaches one can take to analyzing sentiment of text data, and I will primarily use models from the NLTK library. Here a message will be categorized into one of two categories:

* Positve
* Negative

I will use Naive Bayes classification to determining the sentiment of the messages in the `Message` variable. This will involve using Twitter data with known sentiments to create a classifier.


```python
# Download twitter data and sentiment analysis model
nltk.download('twitter_samples')
nltk.download('average_perceptron_tagger')
nltk.download('wordnet')
```

After downloading the Twitter data from the `nltk` library, I will import it from `nltk.corpus` as `twitter_samples`.

```python
from nltk.corpus import twitter_samples
```

My first step is to instantiate the tweets that have been labeled as positive and negative, in order to create a training set that can be fed into the classifier.

```python
# Instantiate positive tweets
positive_tweets = twitter_samples.strings('positive_tweets.json')
# Instantiate negative tweets
negative_tweets = twitter_samples.strings('negative_tweets.json')
```

I will take a quick peek into the types of tweets and their distributions with indexing and `len()`.

```python
# To view 1st positive tweet
print(positive_tweets[0])

# To view the total number of positive tweets
print("There are total {} positive tweets".format(len(positive_tweets)))
```

```python
# To view 1st negative tweet
print(negative_tweets[0])

# To view the total number of negative tweets
print("There are total {} negative tweets".format(len(negative_tweets)))
```

I can see that the number of positive to negative samples used to assemble the training set is evenly distributed. I need to tokenize the tweets.

```python
positive_tweet_tokens = twitter_samples.tokenized('positive_tweets.json')
# Show the first tokenized positive tweet
print(positive_tweet_tokens[0])

negative_tweet_tokens = twitter_samples.tokenized('negative_tweets.json')
# Show the first tokenized negative tweet
print(negative_tweet_tokens[0])
```

Since I removed HTML tags and URLs in the SMS data, I'll do the same for the tweets. I'll use the same functions for this as I defined above.

```python
# Define lists for preprocessed tokens
positive_cleaned_tweets_list = []
negative_cleaned_tweets_list = []

# Positive tokens
for tweet in positive_tweet_tokens:
    cleaned_tweet = []
    for token in tweet:
        # Remove URL
        url_cleaned = remove_urls(token)
        # Remove HTML
        html_cleaned = remove_html(url_cleaned)
        cleaned_tweet.append(html_cleaned)
    # Add to list
    positive_cleaned_tweets_list.append(cleaned_tweet)

# Negative tokens
for tweet in negative_tweet_tokens:
    cleaned_tweet = []
    for token in tweet:
        # Remove URL
        url_cleaned = remove_urls(token)
        # Remove HTML
        html_cleaned = remove_html(url_cleaned)
        cleaned_tweet.append(html_cleaned)
    # Add to list
    negative_cleaned_tweets_list.append(cleaned_tweet)


# View cleaned tokens lists
print(positive_cleaned_tweets_list[:5])
print(len(positive_cleaned_tweets_list))
print(negative_cleaned_tweets_list[:5])
print(len(negative_cleaned_tweets_list))
```

Now I need to select the features for my classifier. I will select the most common words in my corpus because I assume that these words are the best indicators if something is positive or negative. I will use a frequency distribution to select them.

```python
# Create a list of all tweets
list_of_all_tweets = positive_cleaned_tweets_list + negative_cleaned_tweets_list

# Convert my list of lists into a flat list of tokens
all_tweets_tokens = [token for sublist in list_of_all_tweets for token in sublist]

# To check how many word that is
print(len(all_tweets_tokens))
```

```python
# Create a frequency distribution of all the words
all_tokens = nltk.FreqDist(token for token in all_tweets_tokens)

# To inspect the result
print(len(all_tokens))
print(all_tokens.most_common(10))

# Select the top 10,000 words to be our features
sentiment_features = [word for (word, freq) in all_tokens.most_common(10000)]

# To check what our features looks like
print(sentiment_features[:5])
```

Now I need to create a feature extractor that identifies when a feature is present in a message.

```python
def document_features(document , sentiment_features):
  # Make a set of all tokens in a message
  document_tokens = set(document)
  # Create a dictionary of features
  features = {}
  # For all the features, check if the feature is present in my target message
  for token in sentiment_features:
    # Make a dictionary for each message
    features[token] = (token in document_tokens)
  return features

# Now, test it on a simple message / sentence
test_sentence = "coding is fun when you don't have time :)".split(" ")
print(document_features(test_sentence , sentiment_features))
```

Now I need to create training and testing sets, and label them accordingly. 70% of the tweets will be used for the training set, and 30% for the testing set. Using the `random` library, I can randomly select the tweets for each.

```python
import random

# Now, create positive and negative dataset from my list of tokenized tweets
positive_labeled_tw = [(tweet , "Positive") for tweet in positive_cleaned_tweets_list]
negative_labeled_tw = [(tweet , "Negative") for tweet in negative_cleaned_tweets_list]


# Now, create the combined dataset
all_labeled_tweets = positive_labeled_tw + negative_labeled_tw

# Initialize the random number, so that I get the same result every time
random.seed(42)

# Shuffle the order of dataset and random method is applied in place
random.shuffle(all_labeled_tweets)

# Now, create a list of (token, sentiment) pairs for all the features in tweets
feature_set = [(document_features(d, sentiment_features), c) for (d, c) in all_labeled_tweets]

train_set , test_set = feature_set[:7000], feature_set[7000:]
```

I will use the `NaiveBayesClassifier` model with my training and testing sets to build my sentiment model. I will use the `classify()` function to test the performance of my model on my training set.


```python
# Import sentiment analysis and classify model
from nltk import classify
from nltk import NaiveBayesClassifier

# Instantiate a classifier and train on the training set
classifier = NaiveBayesClassifier.train(train_set)

# Evaluate the model with classify function
print(f"Accuracy of the model is: {classify.accuracy(classifier, test_set) * 100:.2f}")

# Show the 20 most informative features for prediction
print(classifier.show_most_informative_features(20))
```

Now I will use my model to classify the sentiment of my messages. I will try it out on just one before applying it to the whole corpus.

```python
# Selecting one sample message to evaluate
sample_message = str(dataset['Message_tokenized'][49])
print(sample_message)

# Printing the prediction of sentiment
print(classifier.classify(dict([token, True] for token in sample_message)))
```

The sentiment of the 50th message (`"yup hey then one day on fri we can ask miwa and jiayin take leave go karaoke"`) is Negative per my model.

Now I will evaluate all the messages in the NUS SMS Corpus.

```python
# Creating an empty list
sentiments = []

# Iterating through the column and predict each response's sentiment, and appending sentiment to new list
for message in dataset['Message_tokenized']:
  sentiments.append(str((classifier.classify(dict([token, True] for token in message)))))

# Adding the list back to our dataframe
dataset['Sentiment'] = sentiments

dataset['Sentiment'].value_counts()
```

Now I want to compare sentiment by country. To make this easier, I will transform my sentiment into a numeric variable. I will assign Positive messages '1' and Negative messages '-1'.


```python
# Defining a function to convert sentiment into binary values
def convert_sentiment(sentiment):
  # Taking in sentiments, and converting them into binary values
  if sentiment == 'Positive':
    return 1
  else:
    return -1

# Creating new feature based on the values returned from function
dataset['Sentiment_score'] = dataset['Sentiment'].apply(convert_sentiment)

# Checking
dataset.head()
```

Now I will group by country and use the `.mean()` method to get the average sentiment score for each country.

```python
# Creating a new dataframe grouped by countries and aggregated by the mean
dataset_sentiments = dataset.groupby(['country']).mean(numeric_only=True)

# Resetting the index
dataset_sentiments.reset_index(inplace = True)

# Checking
dataset_sentiments.head()
```

I want to know which country sends the most positive messages, so I will use the `.sort_values()` method.

```python
# Sort dataframe by Sentiment_score
dataset_sentiments.sort_values(by = 'Sentiment_score')
```

Now I will plot the sentiment by country to see which countries are the most Positive/Negative.

```python
# Plot general sentiments
fig, ax = plt.subplots(figsize = (40, 15))
sns.barplot(x = 'country' , y = 'Sentiment_score' , data = dataset_sentiments , color = 'lightseagreen')
```
